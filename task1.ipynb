```python
# Select only numeric columns
numeric_cols = df.select_dtypes(include=['number']).columns

# Fill missing values in numeric columns with the median of each column
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

# Now, print the first few rows to check
df.head()

```

```python
# Calculate the average score in math (G3)
average_score = df['G3'].mean()
average_score
```

```python
# Count students with final grade greater than 15
students_above_15 = df[df['G3'] > 15].shape[0]
students_above_15
```

```python
# Calculate the correlation between study time and final grade (G3)
correlation = df['studytime'].corr(df['G3'])
correlation
```

```python
# Calculate the average final grade (G3) by gender
average_grade_by_gender = df.groupby('sex')['G3'].mean()
average_grade_by_gender
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Plot histogram for final grades (G3)
plt.hist(df['G3'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Final Grades (G3)')
plt.xlabel('Final Grade (G3)')
plt.ylabel('Number of Students')
plt.show()
```

```python
# Scatter plot between study time and final grade (G3)
plt.scatter(df['studytime'], df['G3'], color='orange')
plt.title('Study Time vs Final Grade (G3)')
plt.xlabel('Study Time (hours/week)')
plt.ylabel('Final Grade (G3)')
plt.show()
```

```python
# Bar chart comparing average scores by gender
average_grade_by_gender.plot(kind='bar', color=['blue', 'pink'])
plt.title('Average Final Grade (G3) by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Final Grade (G3)')
plt.show()
```
